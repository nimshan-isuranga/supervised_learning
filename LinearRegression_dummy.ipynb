```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
data=pd.read_csv("GPA_data.CSV")
```

```python
data.head()
```

```python
data[["Gender","Extra_Curricular"]]
```

```python
df_dum=pd.get_dummies(data[["Gender","Extra_Curricular"]])
df_dum
```

```python
df_dum.drop(["Gender_Female","Extra_Curricular_No"],axis=1,inplace=True) #axis =1 columns
df_dum
```

```python
data_new=pd.concat([data,df_dum],axis=1)
data_new.head()
```

```python
#reindex function can also be used 
```

```python
data_new.drop(["ID","Gender","Extra_Curricular"],axis=1,inplace=True)
data_new.head()
```

```python
x=data_new.iloc[:,data_new.columns !="GPA"].values
```

```python
y=data_new.iloc[:,4].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.score(x_train,y_train)*100
```

```python
y_pred=model.predict(x_test)
```

```python
MSE=mean_squared_error(y_pred,y_test)
MSE
```

```python
RMSE=np.sqrt(MSE)
RMSE
```
